## Resumen y frmulas de regression lineal

### Regresion Lineal:

$H_0(X) =  \theta^{T}X  = \theta_0X_0 + \theta_1X_1 + \theta_2X_2 + \theta_3X_3 + ... +  \theta_nX_n $

Parmetros =  $\theta_0 , \theta_1 ,  \theta_2 ... \theta_n$

Funcin de Costo = $J(\theta_0,\theta_1,\theta_2 ...,\theta_n) = \min_{u} f_{objetivo} = \frac{1}{2m} \sum_{i=1}^{m}(y_{i}- \hat y_{i})^{2}$


### Solucin matricial:


$\theta = (X^{T}X)^{-1}X^{T}y$
